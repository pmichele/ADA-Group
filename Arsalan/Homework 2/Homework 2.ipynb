```python
# Import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import string
```

```python
url="https://www.topuniversities.com/sites/default/files/qs-rankings-data/357051_indicators.txt"
r=requests.get(url)
```

```python
jr=r.json()
```

```python
data=jr['data'][0:200]
```

```python
unis=[]
for d in data:
    name = BeautifulSoup(d['uni'],'html.parser')
    rank = BeautifulSoup(d['overall_rank'],'html.parser')
    country = BeautifulSoup(d['location'],'html.parser')
    region = BeautifulSoup(d['region'],'html.parser')
    unis.append({'Name':name.div.text,'Rank':rank,'Country':country,'Region':region})

df=pd.DataFrame(unis)
df=df[['Name','Rank','Country','Region']]
df
```

```python
def createUrl(name):
    name=name.replace("(","").replace(")","")
    urlPre="https://www.topuniversities.com/universities/"
    url=urlPre+str.lower(name).replace(" ", "-").replace("-of-","-").replace(",","")    
    return url
```
