```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
```

<h1>Task 1</h1>

```python
data=pd.read_csv("lalonde.csv")
for i in range(len(data)):
    if(data.iloc[i]["black"]==1):
        data.set_value(i,"race","black")
    elif(data.iloc[i]["hispan"]==1):
        data.set_value(i,"race","hispanic")
    else:
        data.set_value(i,"race","white")
data.drop("black",axis=1,inplace=True)
data.drop("hispan",axis=1,inplace=True)

control=data[data["treat"]==0]
treatment=data[data["treat"]==1]
```

```python
grp = data[["treat","re78"]].groupby("treat")
mean=grp.mean()
std=grp.std()

mean.index=["Didn't participate","Participated"]
std.index=["Didn't participate","Participated"]

mean.rename(columns={"re78":"Average salary in 1978"},inplace=True)
```

```python
mean.plot(kind='bar',title="Comparison of average annual income between program participants",rot=0)
plt.show()
```

```python
mean
```

From the graphs and numbers above for average salaries in 1978, one would naively conclude that the treatment did not improve salaries and in fact made it worse.
