<h1>Task 1</h1>

```python
data=pd.read_csv("lalonde.csv")
```

```python
grp = data[["treat","re78"]].groupby("treat")
mean=grp.mean()
std=grp.std()

mean.index=["Didn't participate","Participated"]
std.index=["Didn't participate","Participated"]

mean.rename(columns={"re78":"Average salary in 1978"},inplace=True)
```

```python
mean.plot(kind='bar',title="Comparison of average between program participants",rot=0)
plt.show()
```
